```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time 
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36"
}
```

```python
all_companies = []
```

```python
for page in range(1,499):

    print(f"Scraping Page {page}...")

    url = f"https://www.ambitionbox.com/list-of-companies?page={page}"

    response = requests.get(url, headers=headers)

    soup = BeautifulSoup(response.text, "html.parser")

    companies = soup.find_all("div", class_="companyCardWrapper")

    for company in companies:

        name = company.find("h2")
        name = name.text.strip() if name else "N/A"

        rating = company.select_one(".rating_text div")
        rating = rating.text.strip() if rating else "N/A"
        
        reviews = company.find("span", class_="companyCardWrapper__ActionCount")
        reviews = reviews.text.strip() if reviews else "N/A"

        all_companies.append({
            "Company": name,
            "Rating": rating,
            "Reviews": reviews
        })

    # Delay to avoid blocking
    time.sleep(2)
```

```python
df = pd.DataFrame(all_companies)

df.tail()
```

```python
df.to_csv("ambitionbox_500_pages.csv", index=False)
```
